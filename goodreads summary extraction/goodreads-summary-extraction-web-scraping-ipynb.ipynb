```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

df  = pd.DataFrame(columns=["Book_title", 'Book_summary'] )

genres = ["art", "biography", "Business", "childrens", "christian", "classics", "comics", "cookbooks", "ebooks", 
         "fantasy", "fiction", "graphic-novels", "historical-fiction", "history", "horror", "memoir","music",
          "mystery","non-fiction", "poetry","psychology", "romance","science", "science-fiction","self-help",
          "sports","thriller", "travel","young-adult" ]

for genre in genres:
    url = f'https://www.goodreads.com/genres/new_releases/{genre}'
    r = requests.get(url).text
    soup = BeautifulSoup(r, "lxml")
    # print(soup.prettify())

    # Getting the summary data of book from its url
    def get_summary_data(link):
        url1  = link
        print(url1)
        r1 = requests.get(url1).text
        soup1 = BeautifulSoup(r1, "lxml")

        # Getting the book summary
        try:
            summary_cls = soup1.find('div', class_ = 'readable stacked')
            span  = summary_cls.find_all('span', style = 'display:none')[0]
            book_summary = span.text

            # getting the book title
            book_title = soup1.find('h1', id = 'bookTitle').text.strip("\n").lstrip()

            del span, url1, r1, soup1, summary_cls
            return book_title, book_summary
        except:
            return False, False


    # Geeting the urls of list of books and updating them to list
    Book_urls = []
    Bookdetails = soup.find_all('div', class_ = 'leftAlignedImage bookBox')
    for book in Bookdetails:
        book_anchors = book.find('a')
        imgs = book.find('img')
        Book_url = 'https://www.goodreads.com' + book_anchors.get('href')
        Book_title = imgs.get('alt')

        Book_urls.append(Book_url)

    Book_summary_list = []

    for item in Book_urls:
        book_title, book_summary = get_summary_data(item)
        Book_summary_list.append((book_title, book_summary))

    for i, j in Book_summary_list:
        df1 = pd.DataFrame({'Book_title': [i], 'Book_summary': [j]})
        df = pd.concat([df, df1], ignore_index = True, axis = 0)

df.to_csv(r"goodreads_summary_data.csv")
```
